```python
import pandas as pd
import numpy as np
import random

def random_name():
    first_names = ["James", "Mary", "Robert", "Patricia", "John", "Jennifer", "Michael", "Linda", "William", "Elizabeth",
                   "David", "Barbara", "Richard", "Susan", "Joseph", "Jessica", "Thomas", "Sarah", "Charles", "Karen",
                   "Christopher", "Nancy", "Daniel", "Lisa", "Matthew", "Margaret", "Anthony", "Betty", "Mark", "Sandra",
                   "Donald", "Ashley", "Steven", "Kimberly", "Paul", "Emily", "Andrew", "Donna", "Joshua", "Michelle",
                   "Kenneth", "Dorothy", "Kevin", "Carol", "Brian", "Amanda", "George", "Melissa", "Edward", "Deborah",
                   "Ronald", "Stephanie", "Timothy", "Rebecca", "Jason", "Laura", "Jeffrey", "Helen", "Ryan", "Sharon",
                   "Jacob", "Cynthia", "Gary", "Kathleen", "Nicholas", "Amy", "Eric", "Shirley", "Stephen", "Angela",
                   "Jonathan", "Anna", "Larry", "Ruth", "Justin", "Brenda", "Scott", "Pamela", "Brandon", "Nicole",
                   "Frank", "Katherine", "Benjamin", "Samantha", "Gregory", "Christine", "Raymond", "Catherine", 
                   "Samuel", "Virginia", "Patrick", "Debra", "Alexander", "Rachel", "Jack", "Janet", "Dennis", "Emma"]
    last_names = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis", "Rodriguez", "Martinez",
                  "Hernandez", "Lopez", "Gonzalez", "Wilson", "Anderson", "Thomas", "Taylor", "Moore", "Jackson", "Martin",
                  "Lee", "Perez", "Thompson", "White", "Harris", "Sanchez", "Clark", "Ramirez", "Lewis", "Robinson",
                  "Walker", "Young", "Allen", "King", "Wright", "Scott", "Torres", "Nguyen", "Hill", "Flores",
                  "Green", "Adams", "Nelson", "Baker", "Hall", "Rivera", "Campbell", "Mitchell", "Carter", "Roberts"]
    return random.choice(first_names) + " " + random.choice(last_names)


def random_email(username):
    """Generate a random email address."""
    domains = ["@gmail.com", "@yahoo.com", "@hotmail.com", "@example.com"]
    return username.replace(" ", ".").lower() + random.choice(domains)

# Number of rows
num_rows = 50000

# Generate synthetic data
new_data = {
    'UserID': range(1, num_rows + 1),
    'NewUser': np.random.choice([True, False], num_rows),
    'Username': [random_name() for _ in range(num_rows)],  # Using real names as usernames
    'Email': [],  # to be filled later
    'Gender': np.random.choice(['Male', 'Female', 'Other'], num_rows),
    'Age': np.random.randint(18, 70, size=num_rows),
    'Height': np.random.randint(150, 200, size=num_rows),  # assuming height in cm
    'Weight': np.random.randint(50, 150, size=num_rows),   # assuming weight in kg
    'ActivityLevel': np.random.choice(['Sedentary', 'Moderately Active', 'Active', 'Very Active'], num_rows),
    'GoalID': np.random.randint(1, 11, size=num_rows),
    'DietPreferenceID': np.random.randint(1, 11, size=num_rows),
    'RegionID': np.random.randint(1, 11, size=num_rows),
    'AllergenID': np.random.randint(1, 11, size=num_rows)
}

# Fill Email based on Username
new_data['Email'] = [random_email(username) for username in new_data['Username']]

# Create DataFrame
new_df = pd.DataFrame(new_data)

# Display the first few rows of the dataframe
new_df.head()
```

```python
import pandas as pd

df=new_df

import mysql.connector
from mysql.connector import Error

try:
    connection = mysql.connector.connect(
        host='hostname',        # your host, usually localhost
        user='username',        # your username
        password='password',    # your password
        database='databasename' # name of the database
    )
    if connection.is_connected():
        db_Info = connection.get_server_info()
        print("Connected to MySQL Server version ", db_Info)
except Error as e:
    print("Error while connecting to MySQL", e)

from sqlalchemy import create_engine

# Create SQLAlchemy engine
engine = create_engine("mysql+mysqlconnector://{user}:{pw}@{host}/{db}"
                       .format(user="username",
                               pw="password",
                               host="hostname",
                               db="databasename"))

# Write data to SQL
df.to_sql('your_table_name', con=engine, if_exists='replace', index=False)
```
